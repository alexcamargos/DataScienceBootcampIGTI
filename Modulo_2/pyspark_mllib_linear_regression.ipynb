```python
# Criando a SparkSession
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('PySpark MlLib').getOrCreate()
spark
```

```python
from pyspark import SparkFiles
```

```python
# Link para acesso ao dataset.
DATASET_WEB_PATH = 'https://raw.githubusercontent.com/alexcamargos/DataScienceBootcampIGTI/main/Modulo_2/dataset/vendas_censo.csv'

# Informando ao PySpark onde buscar o dataset.
spark.sparkContext.addFile(DATASET_WEB_PATH)

# Fazendo download do arquivo para o espaço local.
DATASET_LOCAL_PATH = SparkFiles.get("vendas_censo.csv")
```

```python
df = spark.read.option('header', True) \
               .option('inferSchema', 'True') \
               .option('encoding', 'utf-8') \
               .csv(f'file://{DATASET_LOCAL_PATH}')
```

## Scatter matrix

```python
import pandas as pd
```

```python
# Listando todas as colunas numéricas integer ou double.
columns_numericas = [column[0] for column in df.dtypes if column[1] in ['int','double' ] if not column[0].startswith('Cod_')]

# Selecionando de forma aleatória 80% dos dados para teste. 
sampled_data = df.select(columns_numericas).sample(False, .7, seed=42).toPandas()
```

```python
axs = pd.plotting.scatter_matrix(sampled_data, figsize=(20, 15))
columns_number = len(sampled_data.columns)

for col_num in range(columns_number):
    v = axs[col_num, 0]
    v.yaxis.label.set_rotation(0)
    v.yaxis.label.set_ha('right')
    v.set_yticks(())
    h = axs[columns_number - 1, col_num]
    h.xaxis.label.set_rotation(90)
    h.set_xticks(())
```

## Correlação entre as colunas que contem informações numéricas

```python
# seaborn - Biblioteca para visualização de dados
import seaborn as sns

# Configurando a exibição dos gráficos para o notebook.
sns.set_theme(style='whitegrid', context='notebook')

# Tamanho padrão dos gráficos.
sns.set(rc={'figure.figsize': (30, 12)})

# Correlação entre as colunas que contem informações numéricas.
# https://www.tylervigen.com/spurious-correlations
heatmap = sns.heatmap(sampled_data[columns_numericas].corr(),
                      vmin=-1,
                      vmax=1,
                      annot=True,
                      cmap='BrBG',
                      center=0)

heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=45, ha='right');
```

## Linear Regression

```python
# Importando as bibliotecas para trabalhar com machine learning.
from pyspark.ml.feature import StringIndexer
from pyspark.ml.feature import OneHotEncoder
from pyspark.ml.feature import VectorAssembler

from pyspark.ml.pipeline import Pipeline

from pyspark.ml.regression import LinearRegression
from pyspark.ml.evaluation import RegressionEvaluator
```

```python
# Variável que desejamos prever.
response_variables = 'Valor_anuncio'
```

```python
# Variáveis utilizadas para realizar a previsão.
predictor_variables = ['Metragem', 'Quartos', 'Banheiros', 'Vagas']
```

```python
# Dividindo os dados em dois grupos, treino 75% e teste 25%.
traing_data, test_data = df.randomSplit(weights=[.75, .25], seed=42)
```

## Engenharia de características (Feature engineering)

```python
vector_assembler = VectorAssembler(inputCols=predictor_variables, outputCol='features')
```

```python
predictor_df = vector_assembler.transform(df)
predictor_df = predictor_df.select('features', response_variables)
predictor_df.show(5, truncate=False)
```

## Regressão linear

```python
regression_model = LinearRegression(labelCol=response_variables, featuresCol="features")
```

```python
pipe_line = Pipeline(stages=[vector_assembler, regression_model])
```

```python
model = pipe_line.fit(traing_data)
```

```python
predictions = model.transform(test_data)
predictions_df = predictions.select('features', response_variables, 'prediction')
predictions_df.show(10, truncate=False)
```

```python
lr_model = model.stages[1]
print(f'Interceptor: {lr_model.intercept}')
print(f'coeficientes: {lr_model.coefficients}')
```

```python
model_summary = lr_model.summary
print(f'Raiz quadrada do erro-médio (Root-mean-square deviation): {model_summary.rootMeanSquaredError:.4f}')
print(f'R² (R Squared): {model_summary.r2:.4f}')
```

```python
model_summary.residuals.show()
```

```python
lr_evaluator_r2 = RegressionEvaluator(predictionCol='prediction', labelCol=response_variables, metricName='r2')
lr_evaluator_mae = RegressionEvaluator(predictionCol='prediction', labelCol=response_variables, metricName='mae')
lr_evaluator_rmse = RegressionEvaluator(predictionCol='prediction', labelCol=response_variables, metricName='rmse')
```

```python
print(f'R² (R Squared) para os dados de teste: {lr_evaluator_r2.evaluate(predictions):.4f}')
print(f'Erro médio absoluto (Mean absolute error) para os dados de teste: {lr_evaluator_mae.evaluate(predictions):.2f}')
print(f'Raiz quadrada do erro-médio (Root-mean-square deviation) para os dados de teste: {lr_evaluator_rmse.evaluate(predictions):.2f}')
```
